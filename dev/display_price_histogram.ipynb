```python
import plotly.express as px
import pandas as pd
PATH = f'./export/rooms_5.csv'
CITIES = ['Biarritz', 'Anglet', 'Bayonne']
```

```python
def load_data(nrows=None):
    data = pd.read_csv(PATH)
    if nrows:
        data = data.head(nrows)
    data['license'].fillna("", inplace=True)
    data['has_license'] = data.license.str.len() > 6
    data['city'] = data.city.str.title().str.strip()
    data = data[data.city.isin(CITIES)]
    return data
```

```python
df = load_data()
df.head(2)
```

```python
# remove outliers
df_plot = df.copy()
df_plot = df_plot[(df_plot.rate > df_plot.rate.quantile(0.01)) & (df_plot.rate < df_plot.rate.quantile(0.95))]

fig = px.histogram(df_plot, x='rate', histnorm='percent', nbins=20, color_discrete_sequence=['#e86b6b'])
fig.update_layout({
    'plot_bgcolor': 'rgba(0,0,0,0)',
    'xaxis': {
        'showgrid': False, # thin lines in the background
        'zeroline': True, # thick line at x=0
        'visible': True,  # numbers below
    },
    'yaxis': {
        'showgrid': False, # thin lines in the background
        'zeroline': False, # thick line at x=0
        'visible': False,  # numbers below
    },
    'bargap': 0.03, # gap between bars of adjacent location coordinates
})
```
